```python


x = np.array([[1, 2],
              [3, 5]])
y = np.array([[5, 6],
              [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])

x_plus_y = x + y
print(x_plus_y)

v_plus_w = v + w
print(v_plus_w)

x_minus_y = x - y
print(x_minus_y)

v_minus_w = v - w
print(v_minus_w)

x_times_2 = x * 2
print(x_times_2)

v_times_2 = v * 2
print(v_times_2)

x_square = x**2
print(x_square)

v_square = v**2
print(v_square)

dot_vw = np.dot(v, w)
print(dot_vw)

dot_xv = np.dot(x, v)
print(dot_xv)

dot_xy = np.dot(x, y)
print(dot_xy)

concat_xy_row = np.concatenate((x, y), axis=0)
print(concat_xy_row)

v_col = v[:, np.newaxis]
w_col = w[:, np.newaxis]

concat_vw_col = np.concatenate((v_col, w_col), axis=1)
print(concat_vw_col)

# np.concatenate((x, v), axis=0) ValueError
v_row = v.reshape(1,2)
concat_x_v = np.concatenate((x, v_row), axis=0)
print(concat_x_v)




```

```python
# Matrices
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1.
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A * A^-1 =\n", I)

# 2.
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB =\n", AB)
print("BA =\n", BA)
print("AB == BA?", np.allclose(AB, BA))

# 3.
AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print("(AB)^T =\n", AB_T)
print("B^T A^T =\n", BT_AT)
print("Are they equal?", np.allclose(AB_T, BT_AT))

# 4.
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Using inverse
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("Solution using inverse:\n", X)

# Using np.linalg.solve
X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve:\n", X_solve)
```
