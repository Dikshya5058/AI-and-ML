```python
import pandas as pd
```

```python
# 3.1-Problem_1

df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/bank.csv")

print(df.info())

obj_cols = df.select_dtypes(include='object').columns
print("Object Columns:", obj_cols)

for col in obj_cols:
    print("Unique values of", col, ":", df[col].unique())

print(df.isnull().sum())

numeric_df = df.select_dtypes(exclude='object')

numeric_df.to_csv("banknumericdata.csv", index=False)

new_df = pd.read_csv("banknumericdata.csv")
print(new_df.describe())
```

```python
# 3.1-Problem_2

df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/medical_students_dataset.csv")

df.info()
print(df.isnull().sum())

df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(df.isnull().sum())

print("Number of duplicate rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("Total rows after removing duplicates:", len(df))
```

```python
# 3.2-Problem_1
df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/Titanic-Dataset.csv")

subset_df = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]

first_class_df = subset_df[subset_df['Pclass'] == 1]

mean_fare = first_class_df['Fare'].mean()
median_fare = first_class_df['Fare'].median()
max_fare = first_class_df['Fare'].max()
min_fare = first_class_df['Fare'].min()

print("First-Class Fare Statistics:")
print("Mean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Maximum Fare:", max_fare)
print("Minimum Fare:", min_fare)
```

```python
# 3.2-Problem_2
df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/Titanic-Dataset.csv")

subset_df = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]

first_class_df = subset_df[subset_df['Pclass'] == 1]

null_count = first_class_df['Age'].isnull().sum()
print("Number of null values in Age column:", null_count)

first_class_df_cleaned = first_class_df.dropna(subset=['Age'])

print("Rows after removing null Age values:", len(first_class_df_cleaned))
```
