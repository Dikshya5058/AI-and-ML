```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/Titanic-Dataset.csv")
df.info()
fare_df=df[['Fare']]
print(fare_df.head())
class_age=df[['Age','Pclass']]
print(class_age.head())
survived_gender=df[['Survived','Sex']]
print(survived_gender.head())
```

```python
fare_gt_100=df[df['Fare']>100]
print(f"Fare greater than 100:\n\n{fare_gt_100}\n")
first_class=df[df['Pclass']==1]
print(f"Passenger's class = 1: \n\n{first_class}")
female_under_18=df[(df['Age']<18) & (df['Sex']=='female')]
print(f"Female under 18: \n\n{female_under_18}")
```

```python
embarked_c_or_s=df[df['Embarked'].isin(['C','S'])]
print(embarked_c_or_s)
first_second_class=df[df['Pclass'].isin([1,2])]
print(first_second_class)
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['fare_per_year']=df['Fare']/df['Age']
print(df)
high_fare_age=df[df['fare_per_year']>5]
print(f"Fare Per Year higher than 5: \n{high_fare_age}/n/n")
high_fare_age_srt=high_fare_age.sort_values(by="fare_per_year",ascending=False)
print(high_fare_age_srt)
result=high_fare_age_srt[['Name','fare_per_year']]
print(result)
```

```python
df['fare_per_class']=df['Fare']/df['Pclass']
adult_males=df[(df['Sex']=='male') & (df['Age']>18)]
print(adult_males.head(10))
adult_males_srt=adult_males.sort_values(by="fare_per_class",ascending=False)
print(adult_males_srt)
result=adult_males_srt[['Name','Age','fare_per_class']]
print(result)
print(df['Age'].dtype)
```

```python
total_fare=df['Fare'].sum()
print(f"Total Fare of all Passengers:\n\n{total_fare}\n")
first_class=df[df["Pclass"]==1]
tF_first_class=first_class['Fare'].sum()
print(f"Total Fare of first class passengers:\n\n{tF_first_class}\n")
second_class=df[df["Pclass"]==2]
tF_second_class=second_class['Fare'].sum()
print(f"Total Fare of second class passengers:\n\n{tF_second_class}\n")
third_class=df[df["Pclass"]==3]
tF_third_class=third_class['Fare'].sum()
print(f"Total fare of third class passengers:\n\n{tF_third_class}\n")
three_class_together=[tF_first_class,tF_second_class,tF_third_class]
print(f"Total Fare of all three classes:\n\n{three_class_together}\n")
proportions=[i/total_fare for i in three_class_together]
print(f"Proportions:\n\n{proportions}")

```
