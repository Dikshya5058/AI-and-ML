```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/diabetes_.csv")
df.head()
df.info()
df.isnull().sum()
df.describe()
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols_with_zero:
    # Check for outliers using IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]

    # Use median if outliers exist, else mean
    if len(outliers) > 0:
        replacement = df[col].median()
    else:
        replacement = df[col].mean()

    df[col] = df[col].replace(0, replacement)
X = df.drop("Outcome", axis=1).values
y = df["Outcome"].values

np.random.seed(42)
indices = np.random.permutation(len(X))

train_size = int(0.7 * len(X))
train_idx = indices[:train_size]
test_idx = indices[train_size:]

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def knn_predict_single(x, X_train, y_train, k):
    distances = []

    for i in range(len(X_train)):
        dist = euclidean_distance(x, X_train[i])
        distances.append((dist, y_train[i]))

    distances.sort(key=lambda x: x[0])
    k_neighbors = distances[:k]

    labels = [label for _, label in k_neighbors]
    return max(set(labels), key=labels.count)

def knn_predict(X_test, X_train, y_train, k):
    predictions = []
    for x in X_test:
        predictions.append(knn_predict_single(x, X_train, y_train, k))
    return np.array(predictions)

def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

k = 3
y_pred = knn_predict(X_test, X_train, y_train, k)
acc = accuracy(y_test, y_pred)
print("Accuracy:", acc)

```

```python
def standard_scaler(X):
    return (X - X.mean(axis=0)) / X.std(axis=0)
X_scaled = standard_scaler(X)

X_train_s = X_scaled[train_idx]
X_test_s = X_scaled[test_idx]

y_pred_scaled = knn_predict(X_test_s, X_train_s, y_train, k)
acc_scaled = accuracy(y_test, y_pred_scaled)
acc_scaled
print("Accuracy on original data:", acc)
print("Accuracy on scaled data:", acc_scaled)

```

```python
k_values = range(1, 16)

acc_original_list = []
acc_scaled_list = []
time_original = []
time_scaled = []

for k in k_values:
    start = time.time()
    pred = knn_predict(X_test, X_train, y_train, k)
    time_original.append(time.time() - start)
    acc_original_list.append(accuracy(y_test, pred))

    start = time.time()
    pred_s = knn_predict(X_test_s, X_train_s, y_train, k)
    time_scaled.append(time.time() - start)
    acc_scaled_list.append(accuracy(y_test, pred_s))

plt.figure()
plt.plot(k_values, acc_original_list, label="Original Data")
plt.plot(k_values, acc_scaled_list, label="Scaled Data")
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title("k vs Accuracy")
plt.legend()
plt.show()

plt.figure()
plt.plot(k_values, time_original, label="Original Data")
plt.plot(k_values, time_scaled, label="Scaled Data")
plt.xlabel("k")
plt.ylabel("Time (seconds)")
plt.title("k vs Time Taken")
plt.legend()
plt.show()

#Effect of k on Accuracy:

# Small k → High variance, sensitive to noise

# Large k → More stable but may underfit

# Best performance observed around k = 5–7

# Effect of k on Computational Cost:

# Time increases as k increases
```
